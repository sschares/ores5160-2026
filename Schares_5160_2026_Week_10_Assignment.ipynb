```python
### Q1: Write a function "count_vowels(s)" that takes a string and returns the number of vowels in it (a, e, i, o, u — both upper and lower case).

def count_vowels(s):
  """
  Counts the number of vowels in a string.
  >>> count_vowels('hello')
  2
  >>> count_vowels('HELLO')
  2
  >>> count_vowels('xyz')
  0
  >>> count_vowels('AEIOUaeiou')
  10
  >>> count_vowels('')
  0
  """
  vowels = 'aeiouAEIOU'
  count = 0
  for char in s:
    if char in vowels:
      count += 1
  return count

print(count_vowels('hello'))
print(count_vowels('HELLO'))
print(count_vowels('xyz'))
```

```python
### Q3: Create a function filter_short(words, max_len) that returns a list of only those words whose length is less than or equal to max_len

def filter_short(words, max_len):
      """
      Filter only those words whose length is less than or equal to max_len
      >>> filter_short(['hi', 'hello', 'no', 'yes'], 2)
      ['hi', 'no']
      >>> filter_short(['hi', 'hello', 'no', 'yes'], 3)
      ['hi', 'no', 'yes']
      >>> filter_short(['hi', 'hello', 'no', 'yes'], 5)
      ['hello']
      """
      ### Begin solution
      return [word for word in words if len(word) <= max_len]
      ### End solution

print(filter_short(['hi', 'hello', 'no', 'yes'], 2))
print(filter_short(['hi', 'hello', 'no', 'yes'], 3))
print(filter_short(['hi', 'hello', 'no', 'yes'], 5))

```
